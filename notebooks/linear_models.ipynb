```python
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from activ import load_data, data_normalization
from sklearn.model_selection import train_test_split, cross_val_score
from activ.pipeline import score_clusters
import matplotlib.pyplot as plt
import scipy
```

```python
nmfdata = load_data(uoinmf=True)
```

```python
nmf_bm = nmfdata.biomarkers
nmf_oc = nmfdata.outcomes
plt.figure(figsize=(10,40))
plt.matshow(nmf_oc.T)
```

```python
mat = []
for x in range(nmf_oc.shape[1]):
    y = nmf_oc[:,x]
    regr = linear_model.LinearRegression()
    y_pred = regr.fit(nmf_bm, y).coef_
    mat.append(y_pred)
plt.matshow(mat, cmap='gray_r')
plt.yticks(np.arange(6), ('Sleep', 'Visual/Vestibular', 'Recovered', 'Depression', 'Cognitive', 'PTSD'))
plt.xticks(np.arange(6), ('4','2','3','1','6','5'))
plt.xlabel('Biomarkers')
plt.ylabel('Outcomes')    
```
