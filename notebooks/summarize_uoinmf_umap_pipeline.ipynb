```python
%matplotlib inline
import matplotlib.pyplot as plt
import h5py
import numpy as np
from activ.analytics import heatmap
from activ.clustering import UmapClusteringResults

def print_max(res, op='mean'):
    adj = res.adjust(op=op)
    ui, ci = np.unravel_index(np.argmax(adj, axis=None), adj.shape)
    print("max", op, ":", adj[ui, ci], '-', res.umap_dims[ui], "dimensions,", res.cluster_sizes[ci], "clusters")
```

```python
results = UmapClusteringResults("/home/ahyeon/bm_results/umap_cluster_uoinmf2.h5")
```

```python
## UMAP dimensions: 2-10, Cluster sizes: 2-25
```

```python
print_max(results)
ret = results.make_heatmap()
plt.savefig('mean_heatmap_uoinmf_umap.pdf')
```

```python
print_max(results, op='median')
ret = results.make_heatmap(op='median')
plt.savefig('median_heatmap_uoinmf_umap.pdf')
```
