```python
%matplotlib inline
from activ import load_data
import matplotlib.pyplot as plt
from activ.analytics import nmf_bases_heatmap
from matplotlib.backends.backend_pdf import PdfPages
pdf = PdfPages("nmf_results.pdf")
```

```python
uoinmf_data = load_data('uoinmf')
raw_data = load_data()
```

```python
data = uoinmf_data.outcome_bases
labels = raw_data.outcome_features
plt.figure(figsize=(90, 30))
nmf_bases_heatmap(data, col_labels=labels, sort=True, ax=None,
            cbar_kw={}, cbarlabel="", xlab='Outcome Features', ylab='Factors', aspect='auto',
            title='Outcome Factors')
pdf.savefig()
```

```python
data = uoinmf_data.biomarker_bases
labels = raw_data.biomarker_features
plt.figure(figsize=(90, 30))
nmf_bases_heatmap(data, col_labels=labels, sort=True, ax=None,
            cbar_kw={}, cbarlabel="", xlab='Biomarker Features', ylab='Factors', aspect='auto',
            title='Biomarker Factors')
pdf.savefig()
```

```python
pdf.close()
```
