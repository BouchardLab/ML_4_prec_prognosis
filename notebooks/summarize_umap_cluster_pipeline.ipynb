# Summarize result of parameter sweep across UMAP dimension and Number of outcome clusters

```python
%matplotlib inline
import matplotlib.pyplot as plt
import h5py
import numpy as np
from activ.analytics import heatmap
from activ.clustering import UmapClusteringResults

def print_max(res, op='mean'):
    adj = res.adjust(op=op)
    ui, ci = np.unravel_index(np.argmax(adj, axis=None), adj.shape)
    print("max", op, ":", adj[ui, ci], '-', res.umap_dims[ui], "dimensions,", res.cluster_sizes[ci], "clusters")
    
```

```python
results = UmapClusteringResults("../results/sweep_dims2-10_clus2-14/umap_clustering_sweep.h5")
results10_20 = UmapClusteringResults("../results/sweep_dims2-10_clus10-20/umap_clustering_sweep.h5")
```

## UMAP dimensions: 2-10, Cluster sizes: 2-14

```python
print_max(results)
ret = results.make_heatmap()
```

```python
print_max(results, op='median')
ret = results.make_heatmap(op='median')
```

## UMAP dimensions: 2-10, Cluster sizes: 10-20

```python
print_max(results10_20)
ret = results10_20.make_heatmap()
```

```python
print_max(results10_20, op='median')
ret = results10_20.make_heatmap(op='median')
```
