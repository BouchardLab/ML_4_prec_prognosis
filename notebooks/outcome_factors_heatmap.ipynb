```python
%matplotlib inline
from activ import load_data
import matplotlib.pyplot as plt
from activ.analytics import outcome_factors_heatmap
import numpy as np
import h5py
```

```python
data = load_data('uoinmf')
```

```python
txt = '/Users/ahyeon/Desktop/TBIanalysis/data/feature_name_subset_outcome.txt'
with open(txt) as outcomes:
    outcome_features = [each_line.strip() for each_line in outcomes]
```

```python
order = np.argsort(np.max(data.outcome_bases, axis=0))
sorted_bases = data.outcome_bases[:,order[::-1]]

order = np.argsort(np.max(data.outcome_bases, axis=1))
sorted_bases = sorted_bases[order[::-1],:]
sorted_bases.shape
```

```python
plt.figure(figsize=(40, 10))
outcome_factors_heatmap(sorted_bases, range(1,41), col_labels=outcome_features, ax=None,
            cbar_kw={}, cbarlabel="", xlab='Outcome Features', ylab='Factors',
            title='Outcome Factors Heatmap')
plt.savefig('outcome_factors_heatmap.pdf')
```
