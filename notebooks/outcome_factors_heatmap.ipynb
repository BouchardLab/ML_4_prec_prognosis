```python
%matplotlib inline
from activ import load_data
import matplotlib.pyplot as plt
from activ.analytics import nmf_bases_heatmap
import numpy as np
import h5py
```

```python
uoinmf_data = load_data('uoinmf')
raw_data = load_data()
data = uoinmf_data.outcome_bases
labels = raw_data.outcome_features
```

```python
plt.figure(figsize=(40, 10))
nmf_bases_heatmap(data, range(1,41), col_labels=labels, sort=True, ax=None,
            cbar_kw={}, cbarlabel="", xlab='Outcome Features', ylab='Factors',
            title='Outcome Factors Heatmap')
plt.savefig('outcome_factors_heatmap.pdf')
```
